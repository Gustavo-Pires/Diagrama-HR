Leitura dos dados do arquivo

```python
# Carrega o arquivo Excel
workbook = load_workbook('dados.xlsx')
sheet = workbook.active

# Leitura dos dados do arquivo Excel
data = []
for row in sheet.iter_rows(min_row=2, values_only=True):
    name = row[0]
    temperature = row[1]
    luminosity = row[2]

    if temperature is not None and luminosity is not None:
        data.append((name, float(temperature), float(luminosity)))
```

Funcao que define o mapeamento do tamanho da bolinha baseado na luminosidade

```python
# Mapeamento do tamanho da bolinha baseado no tamanho relativo
def size_mapping(temperature):
    return 10 + 2 * (temperature - 5778)  # Ajuste o fator conforme necessário
```

Criação do diagrama HR

```python
# Criação do diagrama HR
plt.figure(figsize=(10, 6))
for name, temperature, luminosity in data:
    size = size_mapping(temperature)
    plt.scatter(temperature, luminosity, s=size, label=name)

plt.scatter(5778, 1, color='red', s=size_mapping(5778), label='Sol')

plt.gca().invert_xaxis()  # Inverte o eixo X para temperatura decrescente
plt.xlabel('Temperatura (K)')
plt.ylabel('Luminosidade')
plt.title('Diagrama HR')
plt.legend()
plt.grid()

# Salva o gráfico em formato PNG
plt.savefig('diagrama_hr.png', dpi=300, bbox_inches='tight')
```
